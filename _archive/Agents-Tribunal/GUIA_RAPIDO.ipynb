
## 🔄 Como Funciona o Tribunal

Quando fazes uma pergunta, o sistema:

1. **Cria resposta inicial simples**
2. **Observer (Passagem 1)** → Descreve o que vê
3. **Clínico (Passagem 1)** → Interpreta em contexto médico
4. **Engenheiro (Passagem 1)** → Adiciona perspetiva de engenharia
5. **Clínico (Passagem 2)** → Revalida precisão clínica
6. **Observer (Passagem 2)** → Verifica coerência visual
7. **Meta-Tribunal** → Síntese final

**Tu só vês a versão final**, mas todas as iterações ficam guardadas.
